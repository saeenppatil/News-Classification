```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

model_path = "./fine_tuned_bert"
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForSequenceClassification.from_pretrained(model_path)

file_path = "test.txt"
with open(file_path, "r") as file:
    random_article = file.read()

print(f"Loaded article: {random_article}")

inputs = tokenizer(
    random_article,
    max_length=256,
    padding="max_length",
    truncation=True,
    return_tensors="pt"
)

model.eval()
with torch.no_grad():
    outputs = model(**inputs)
    logits = outputs.logits

probabilities = torch.sigmoid(logits)

threshold = 0.3
predictions = (probabilities > threshold).int()

categories = ['Politics', 'Business', 'Health', 'Science', 'Technology', 
              'Sports', 'Entertainment', 'World', 'Society', 'Everyday']

predicted_categories = [categories[i] for i in range(len(categories)) if predictions[0][i] == 1]

print(f"Predicted categories: {predicted_categories}")
```

```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

model_path = "./fine_tuned_bert2"
tokenizer = AutoTokenizer.from_pretrained(model_path)
model = AutoModelForSequenceClassification.from_pretrained(model_path)

file_path = "test.txt"
with open(file_path, "r") as file:
    random_article = file.read()

print(f"Loaded article: {random_article}")

inputs = tokenizer(
    random_article,
    max_length=256,
    padding="max_length",
    truncation=True,
    return_tensors="pt"
)

model.eval()
with torch.no_grad():
    outputs = model(**inputs)
    logits = outputs.logits

probabilities = torch.softmax(logits, dim=-1)

predicted_class_index = torch.argmax(probabilities, dim=-1).item()

categories = [
    'crime, law and justice', 'arts, culture, entertainment and media', 'economy, business and finance', 
    'disaster, accident and emergency incident', 'environment', 'education', 'health', 'human interest', 
    'lifestyle and leisure', 'politics', 'labour', 'religion and belief', 'science and technology', 'society', 
    'sport', 'conflict, war and peace', 'weather'
]

# Map the predicted index to the corresponding category
predicted_category = categories[predicted_class_index]

print(f"Predicted category: {predicted_category}")
```
